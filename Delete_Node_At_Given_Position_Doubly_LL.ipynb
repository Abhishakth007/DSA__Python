```python
class Node():
    
    def __init__(self,data , prev = None , next = None):
        
        self.data = data
        
        self.prev = prev
        
        self.next = next
        
class Linked_List():
    
    def __init__(self,head = None):
        
        self.head = head
        
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node 
            
            return
        
        node.next = self.head
        
        self.head.prev = node
        
        self.head = node
        
    def display_linked_list(self):
        
        if self.head is None:
            
            print("Empty LL")
            
            return
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data , end = " --> ")
            
            current_node = current_node.next
            
        print("Null")
        
    def delete_at_position(self,x):
        
        if x is None or x < 0 or self.head is None:
            
            return -1
        
        current_node = self.head
        
        if x == 1:
            
            self.head = current_node.next
            
            if self.head is not None:
                
                    current_node.prev = None
                    
        current_pos = 1
                    
        while current_node is not None and current_pos < x:
            
            current_node = current_node.next
            
            current_pos+=1
            
        
        if current_node is not None:
            
            if current_node.next is not None:
                
                current_node.prev.next = current_node.next
            
            if current_node.prev is not None:
                
                current_node.next.prev = current_node.prev
                
        return                                                   # Time = O(n)......Space = O(1)
    
    
    def insert_multiple(self,list):                      
        
        for values in list:
            
            self.insertion(values)
    
ll = Linked_List()

ll.insert_multiple([5,4,3,2,1])

ll.display_linked_list()

ll.delete_at_position(3)

ll.display_linked_list()
                    
       
        
        
```
