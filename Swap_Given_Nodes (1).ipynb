```python
class Node():
    
    def __init__(self, data , next_node = None):
        
        self.data = data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head = None):
        
        self.head = head
        
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
        
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
            
    def display_linked_list(self):
        
        if self.head is None:
            
            print("Empty linked list")
            
            return
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data , end = " --> ")
            
            current_node = current_node.next_node
            
        print("Null")
        
        
    def swap_nodes(self,x,y):
        
        if x == y:
            
            return 
                
        x_pointer = self.head
        
        y_pointer = self.head
        
        x_prev = None
        
        y_prev = None
        
        while x_pointer is not None and x_pointer.data != x:
            
            x_prev = x_pointer
            
            x_pointer = x_pointer.next_node
            
        x_ahead = x_pointer.next_node
            
            
   
        while y_pointer is not None and y_pointer.data != y:
            
            y_prev = y_pointer
            
            y_pointer = y_pointer.next_node
            
        y_ahead = y_pointer.next_node
        
        
        
        if not x_pointer or not y_pointer:
            
            return
        
        
        if x_prev:
            
            x_prev.next_node = y_pointer
            
        else:
            
            self.head = y_pointer
        
        if y_prev:
            
            y_prev.next_node = x_pointer
            
        else:
            
            self.head = x_pointer
     
        
        x_pointer.next_node , y_pointer.next_node =  y_pointer.next_node , x_pointer.next_node

        
               
            
        
ll = Linked_List()          # Time = O(n)  , Space = O(1)

ll.insert_multiple([8,7,6,5,4,3,2,1])

ll.display_linked_list()

ll.swap_nodes(1,8)

ll.display_linked_list()
```
