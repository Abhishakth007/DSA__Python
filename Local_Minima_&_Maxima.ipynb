```python
def minima_and_maxima(arr):
    
    local_minima= []
    
    local_maxima = []
    
    n = len(arr)-1
    
    if n == 1:
        return [[-1], [-1]]
    
    if n == 2:
        
        if arr[0] < arr[1]:
            
            return [[0], [1]]
        
        elif arr[0] > arr[1]:
            
            return [[1], [0]]
        
        else:
            
            return [[-1], [-1]]
    
    if arr[0]>arr[1]:
        
        local_maxima.append(0)
    else:
        
        local_minima.append(0)
    
        
    
    for i in range(1,len(arr)-1):
        
        prev_element = i-1
        
        current_element = i
        
        next_element = i+1
        
        if arr[current_element] > arr[prev_element] and arr[current_element] > arr[next_element]:
            
            local_maxima.append(i)
            
        elif arr[current_element] < arr[prev_element] and arr[current_element] < arr[next_element]:
            
            local_minima.append(i)
            
            
    if arr[-1] > arr[-2]:
        
        local_maxima.append(n)
    else:
        local.minima.append(n)
            
    
    final_result = []
            
    
    if len(local_minima) == 0:
        
        final_result.append([-1])
        
    else:
        
        final_result.append(local_minima)
        
    if len(local_maxima) == 0:
        
        final_result.append([-1])
        
    else:
        
        final_result.append(local_maxima)
        
    return final_result


nums = [10 ,5, 20, 30, 40, 35, 50]    # Time = O(n)..Space = O(n)
        
result = minima_and_maxima(nums)

print(result)
            
```
