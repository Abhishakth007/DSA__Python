```python
class Node():
    
    def __init__(self,data,next_node = None):
        
        self.data = data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head = None):
        
        self.head = head
        
    
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
        
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
    
    def display_linked_list(self):
        
        if self.head is None:
            
            print("Empty Linked List")
            
            return 
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data , end = " --> ")
            
            current_node = current_node.next_node
            
        print("Null")
        
        
    def remove_duplicates(self):
        
        if self.head is None:
            
            print("Empty Linked List")
            
            return
        
        prev_node = None
        
        current_node = self.head
        
        seen = set()
        
        while current_node is not None:
            
            if current_node.data in seen:
                
                prev_node.next_node = current_node.next_node
                
            else:
                
                seen.add(current_node.data)
                
                prev_node = current_node
                
            current_node = current_node.next_node
            
        return
            
                          
        
ll = Linked_List()

ll.insert_multiple([8,8,7,6,5,4,4,4,4,4,4,4,3,2,1,1])      # Time = O(n)  ..... Space = O(n)

ll.display_linked_list()

ll.remove_duplicates()

ll.display_linked_list()
        
```

```python
# Another Approach

class Node():
    
    def __init__(self,data,next_node = None):
        
        self.data = data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head = None):
        
        self.head = head
        
    
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
        
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
    
    def display_linked_list(self):
        
        if self.head is None:
            
            print("Empty Linked List")
            
            return 
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data , end = " --> ")
            
            current_node = current_node.next_node
            
        print("Null")
        
        
    def remove_duplicates(self):
        
        if self.head is None:
            
            print("Empty Linked List")
            
            return
        
        prev_node = self.head
        
        current_node = self.head.next_node
                
        while current_node is not None:
            
            if prev_node.data == current_node.data:
                
                prev_node.next_node = current_node.next_node
                 
                
            else:
                                
                prev_node = current_node
                
            current_node = current_node.next_node
            
        return
            
                          
        
ll = Linked_List()

ll.insert_multiple([2,2,2,2,2,2])

                                                      # Time = O(n)  ..... Space = O(1)

ll.display_linked_list()

ll.remove_duplicates()

ll.display_linked_list()
        
```
