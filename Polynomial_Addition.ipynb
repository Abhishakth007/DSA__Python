```python
class Node():
    
    def __init__(self,coef, power,next = None):
        
        self.coef = coef
        
        self.power = power
        
        self.next = next
        
class Linked_List():
    
    def __init__(self,head = None):
        
        self.head = head
        
    
    def add_polynomial(self , poly1_head , poly2_head):
        
        current_node1 = poly1_head
        
        current_node2 = poly2_head
        
        new_ll = Node(0,0)
        
        current = new_ll
        
        new_ll_start = current
        
        while current_node1 is not None and current_node2 is not None:
            
            if current_node1.power == current_node2.power:
                
                current.next = Node(current_node1.coef + current_node2.coef, current_node1.power)
                
                current_node1 = current_node1.next
                
                current_node2 = current_node2.next
                
            elif current_node2.power > current_node1.power:
                
                current.next = Node(current_node2.coeff, current_node2.power)
                
                current_node2 = current_node2.next
                
            elif current_node1.power > current_node2.power:
                
                current.next = Node(current_node1.coef, current_node1.power)
                
                current_node1 = current_node1.next
                
            current = current.next
            
        while current_node1 is not None:
            
            current.next = Node(current_node1.coef,current_node1.power)
            
            current_node1 = current_node1.next
            
        while current_node2 is not None:
            
            current.next = Node(current_node2.coef,current_node2.power)
            
            current_node2 = current_node2.next
            
        return new_ll_start.next
    
    def display_linked_list(self,start): 
        
        if start is None:
            
            print("Empty Linked List")
            
            return
        
        current_node = start
        
        while current_node is not None:
            
            if current_node.next is None :
                
                print(f"{current_node.coef}^{current_node.power} ", end = "")
            
            else:
                
                print(f"{current_node.coef}^{current_node.power} ",end = " + ")
            
            current_node = current_node.next
                    
        
poly1 = Linked_List()

poly1.head = Node(5,2,Node(4,1,Node(2,0)))  # Time : O(n + m) .... Space : O(max(n, m))


poly2 = Linked_List()

poly2.head = Node(-5,1)

poly2.head.next = Node(-5,0)

ll = Linked_List()


ll.display_linked_list(ll.add_polynomial(poly1.head,poly2.head)
)
```
