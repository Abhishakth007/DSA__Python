```python
class Node():
    
    def __init__(self,data,next_node = None):
        
        self.data = data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head=None):
        
        self.head = head
        
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
        
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
            
            
    def search_for_an_element(self,k):
        
        if self.head is None:
            
            print("Empty_Linked List")
            
            return
        
        position =0
        
        current_node = self.head
        
        while current_node is not None:
            
            if current_node.data == k:
                
                print("The element is at index : ",position)
                
                return
            
            current_node = current_node.next_node
            
            position+=1
            
        else:
            
            print("No Such Element Found")
            
            return
            
        
    def display_linked_list(self):
        
        if self.head is None :
            
            print("Empty Linked List")
            
            return
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data,end = ' --> ')
            
            current_node = current_node.next_node
            
        print("Null")
        
        return                                          # Time = O(n) ... Space = O(1)
    

ll = Linked_List()

ll.insert_multiple([1,2,3,4])

ll.display_linked_list()

ll.search_for_an_element(1)
   
        
            
            
```

```python
# Recursive Approach

class Node():
    
    def __init__(self,data,next_node = None):
        
        self.data = data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head=None):
        
        self.head = head
        
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
        
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
            
            
    def search_for_an_element(self,start_node , k , position = 0 ):
        
        if start_node is None:
            
            print("Empty_Linked List")
            
            return
                
        if start_node.data == k:
            
            print("The Element is at index : ",position)
            
            return
                
        
        return self.search_for_an_element(start_node.next_node , k ,position+1)
            
        
    def display_linked_list(self):
        
        if self.head is None :
            
            print("Empty Linked List")
            
            return
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data,end = ' --> ')
            
            current_node = current_node.next_node
            
        print("Null")
        
        return                                          # Time = O(n) ... Space = O(n) [Stack will be used for recursion]
    

ll = Linked_List()

ll.insert_multiple([1,2,3,4])

ll.display_linked_list()

ll.search_for_an_element(ll.head, 1)
```
