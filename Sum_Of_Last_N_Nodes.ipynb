```python
class Node():
    
    def __init__(self,data,next_node = None):
        
        self.data = data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head = None):
        
        self.head = head
        
    
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
        
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
    
    def display_linked_list(self):
        
        if self.head is None:
            
            print("Empty Linked List")
            
            return 
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data , end = " --> ")
            
            current_node = current_node.next_node
            
        print("Null")
        
        
    def sum_of_last_n_nodes(self,n):
        
        if self.head is None:
            
            print("Empty Linked List")
            
            return
        
        sum_calculator = 0
        
        behind_pointer = self.head
        
        ahead_pointer = self.head
        
        for i in range(n-1):
            
            if ahead_pointer.next_node is None:
                
                break
            
            ahead_pointer = ahead_pointer.next_node
            
        while ahead_pointer is not None and ahead_pointer.next_node is not None:
            
            behind_pointer = behind_pointer.next_node
            
            ahead_pointer = ahead_pointer.next_node
            
        while behind_pointer is not None:
            
            sum_calculator += behind_pointer.data
        
            behind_pointer = behind_pointer.next_node
            
            
        return sum_calculator
            
        
        
                
        
ll = Linked_List()

ll.insert_multiple([8,7,6,5,4,3,2,1])      # Time = O(n)  ..... Space = O(1)

ll.display_linked_list()

ll.sum_of_last_n_nodes(3)




        
```
