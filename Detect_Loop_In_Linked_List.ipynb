```python
# Another Approach

class Node():
    
    def __init__(self,data,next_node = None):
        
        self.data =data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head =None):
        
        self.head =head
        
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
    
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
            
    def create_loop(self):
        
        loop_position = 0
        
        current_node = self.head
        
        loop_start_pointer = None
                
        while current_node is not None:
            
            if loop_position == 3:
                
                loop_start_pointer = current_node
                
            if loop_position == 6:
                
                current_node.next_node = loop_start_pointer
                
                return
                
            
            current_node = current_node.next_node
            
            loop_position +=1
            
    def detect_loop(self):
        
        slow_ptr = self.head
        
        fast_ptr = self.head
        
        while fast_ptr.next_node is not None and fast_ptr is not None:
            
            slow_ptr = slow_ptr.next_node
            
            fast_ptr = fast_ptr.next_node.next_node
            
            if slow_ptr == fast_ptr:
                
                return True
            
            
        return False
            
    
              
            
    def display_linked_list(self):
        
        if self.head is None:
            
            print("Empty")
            
            return
        
        visited = set()
        
        current_node = self.head
        
        while current_node is not None:
            
            if current_node in visited:
                
                break
            
            print(current_node.data, end = " --> ")
            
            visited.add(current_node)
            
            current_node = current_node.next_node
            
        print("Null")
            
        
        
                                                         # Time = O(n)....Space = O(n)
        

ll = Linked_List()

ll.insert_multiple([17,15,8,12,10,5,4,1,7,6])

ll.display_linked_list()

ll.create_loop()


ll.detect_loop()
            
            
```
