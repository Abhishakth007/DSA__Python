```python
class Node():
    
    def __init__(self,data, next = None):
        
        self.data = data
        
        self.next = next
        
class Linked_List():
    
    def __init__(self,head = None):
        
        self.head = head
        
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next = self.head
        
        self.head = node
        
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
            
        return self
            
    def display_linked_list(self):
        
        if self.head is None:
            
            print("Empty Linked List")
            
            return
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data,end = " --> ")
            
            current_node = current_node.next
            
        print("Null")
        
    
    def add_two_linked_lists(self,ll1,ll2):
        
        ll1_sum = 0
        
        ll2_sum = 0
        
        current_node1 = ll1.head
        
        current_node2 = ll2.head
        
        multiplier = 1
        
        while current_node1 is not None:
            
            ll1_sum += current_node1.data * multiplier
            
            multiplier *= 10
            
            current_node1 = current_node1.next
            
        multiplier =1
            
        while current_node2 is not None:
            
            ll2_sum += current_node2.data * multiplier
            
            multiplier *= 10
            
            current_node2 = current_node2.next
            
        
                    
        resultant_sum = ll1_sum + ll2_sum
        
        
        result_str = str(resultant_sum)[::-1]
        
        return self.insert_multiple(result_str)     # Time = O(n+m)  ... space = O(n+m)
            
            
        
ll1 = Linked_List().insert_multiple([5,6,3])

ll2 = Linked_List().insert_multiple([8,4,2])

ll1.display_linked_list()

ll2.display_linked_list()

ll3 = Linked_List().add_two_linked_lists(ll1,ll2)

ll3.display_linked_list()
```
