```python
class Node():
    
    def __init__(self, data , next_node = None):
        
        self.data = data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head = None):
        
        self.head = head
        
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
        
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
            
    def display_linked_list(self):
        
        if self.head is None:
            
            print("Empty linked list")
            
            return
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data , end = " --> ")
            
            current_node = current_node.next_node
            
        print("Null")
        
        
    def reverse_the_linked_list_from_mid(self):
        
        slow_ptr = self.head
        
        fast_ptr = self.head
        
        while fast_ptr and fast_ptr.next_node :
            
            
            slow_ptr_prev = slow_ptr
            
            
            slow_ptr = slow_ptr.next_node

            
            fast_ptr = fast_ptr.next_node.next_node
                
        
        prev_node = None
        
        current_node = slow_ptr
        
        while current_node is not None:
            
            nxt_node = current_node.next_node
            
            current_node.next_node = prev_node
            
            prev_node = current_node
            
            current_node = nxt_node
            
        slow_ptr_prev.next_node = prev_node
        
        return prev_node
        
    def check_if_palindrome(self):
        
        if self.head is None or self.head.next_node is None:
            
            return 
                 
        current_node = self.head 
            
        reversed_second_half_head = self.reverse_the_linked_list_from_mid()
            
        mid_ptr = reversed_second_half_head
            
        while current_node is not mid_ptr:
                
            if current_node.data == reversed_second_half_head.data:
                    
                current_node = current_node.next_node
                    
                reversed_second_half_head = reversed_second_half_head.next_node
                    
            else:
                    
                print("Not Palindrome")
                    
                return 
                    
        print("Yes Palindrome")
                               
        
ll = Linked_List()                                    # Time = O(n)  , Space = O(1)

ll.insert_multiple(['a', 'b', 'a', 'c', 'a', 'b', 'a'])

ll.display_linked_list()

ll.check_if_palindrome()

```
