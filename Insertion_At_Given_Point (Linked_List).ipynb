```python
class Node():
    
    def __init__(self,data,next_node = None):
        
        self.data = data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head = None):
        
        self.head = head
        
    def insertion_at_given_point(self , value , pointer):
        
        node = Node(value)
        
        if self.head is None:
            
            print("Linked List Is Empty")
            
            return
        
        current_node = self.head
        
        while current_node is not None:
            
            if current_node.data == pointer:
                
                node.next_node = current_node.next_node
                
                current_node.next_node = node
                                
                return
            
            else:
                current_node = current_node.next_node
                
        else:
            
            print("No Such Element Found")
            
    def insert_at_the_beginning(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
        
    
    
    def display_the_linked_list(self):
        
        if self.head is None:
            
            print("No Elements in the Linked List")
                        
            return
        
        current_node = self.head
        
        while True:
            
            print(current_node.data , end = " --> ")
            
            if current_node.next_node is None:
                
                print("Null")
                
                break
                
                return

            
            current_node = current_node.next_node   # (For Overall Operations) Auxiliary Time = O(n)  , Auxiliary Space = O(1)... 
                                                     
                                                    #Insertion after a give pointer..
            
ll = Linked_List()

ll.insert_at_the_beginning(1)

ll.insert_at_the_beginning(2)

ll.insert_at_the_beginning(3)

ll.display_the_linked_list()

ll.insertion_at_given_point(10,1)

ll.display_the_linked_list()



   
        
```
