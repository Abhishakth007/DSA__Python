```python
class Node():
    
    def __init__(self,data,next_node = None):
        
        self.data =data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head =None):
        
        self.head =head
        
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
    
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
            
    def segregate_even_and_odd(self):
        
        even_arr = []
        
        odd_arr = []
        
        current_node = self.head
        
        while current_node is not None:
            
            if current_node.data % 2 == 0:
                
                even_arr.append(current_node.data)
                
            else:
                
                odd_arr.append(current_node.data)
                
            current_node = current_node.next_node
            
        self.head = None
        
        self.insert_multiple(odd_arr)
            
        self.insert_multiple(even_arr)
        
                
                
         
                
            
    def display_linked_list(self):
        
        if self.head is None:
            
            print("Empty")
            
            return
        
        
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data,end = " --> ")
            
            current_node = current_node.next_node
            
        print("Null")
                                                         # Time = O(n)....Space = O(n)
        

ll = Linked_List()

ll.insert_multiple([17,15,8,12,10,5,4,1,7,6])

ll.segregate_even_and_odd()

ll.display_linked_list()
            
            
```
