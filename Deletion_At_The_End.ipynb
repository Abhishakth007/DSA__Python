```python
class Node():
    
    def __init__(self,data,next_node = None):
        
        self.data = data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head = None):
        
        self.head = head
        
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head =  node
        
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
        
    def delete_from_end(self):
        
        current_node = self.head
        
        while True:
            
            if current_node.next_node.next_node is None:
                
                current_node.next_node= None
                
                return
                
            
            current_node = current_node.next_node
        
    def display_linked_list(self):
        
        current_node = self.head
        
        while True:
            
            print(current_node.data,end =" --> ")
            
            if current_node.next_node is None:
                
                print("Null")
                
                return
            
            current_node = current_node.next_node      # Worst Case...Time = O(n)  , Space = O(1)
            

ll = Linked_List()

ll.insert_multiple([1,2,3,4])

ll.display_linked_list()

ll.delete_from_end()

ll.display_linked_list()

           
```
