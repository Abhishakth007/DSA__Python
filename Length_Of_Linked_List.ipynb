```python
class Node():
    
    def __init__(self,data,next_node = None):
        
        self.data = data
        
        self.next_node = next_node
        
class Linked_List():
    
    def __init__(self,head=None):
        
        self.head = head
        
    def insertion(self,value):
        
        node = Node(value)
        
        if self.head is None:
            
            self.head = node
            
            return
        
        node.next_node = self.head
        
        self.head = node
        
    def insert_multiple(self,list):
        
        for values in list:
            
            self.insertion(values)
            
            
    def total_no_of_nodes(self):
        
        length = 0
        
        if self.head is None:
            
            print("No Elements in the linked list")
            
            return
        
        current_node = self.head
        
        while current_node is not None:
            
            length+=1
            
            current_node = current_node.next_node
            
            if current_node.next_node is None:
                
                print("No_of_Nodes = ", length+1)
                
                return 
            
        
    def display_linked_list(self):
        
        if self.head is None :
            
            print("Empty Linked List")
            
            return
        
        current_node = self.head
        
        while current_node is not None:
            
            print(current_node.data,end = ' --> ')
            
            current_node = current_node.next_node
            
        print("Null")
        
        return                                          # Time = O(n) ... Space = O(1)
    

ll = Linked_List()

ll.insert_multiple([1,2,3,4])

ll.display_linked_list()

ll.total_no_of_nodes()
```
